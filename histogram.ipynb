```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('emotic_radiant_fog_160_dev.csv')

# Count the occurrences of each category and calculate percentages
category_counts = df['category'].value_counts()
category_percentages = df['category'].value_counts(normalize=True) * 100

# Create a subplot with a shared x-axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the count numbers
category_counts.plot(kind='bar', color='skyblue', label='Count', ax=ax)

# Calculate the width for tighter layout
width = 0.4

# Plot the percentages on a secondary y-axis
ax2 = ax.twinx()
ax2.plot(ax.get_xticks(), category_percentages, color='orange', marker='o', label='Percentage (%)')

# Set x-axis labels with smaller distance between them
ax.set_xticks(range(len(category_counts)))
ax.set_xticklabels(category_counts.index, rotation=45, fontsize=10)

plt.title('Category Histogram')
ax.set_xlabel('Category')
ax.set_ylabel('Count')
ax2.set_ylabel('Percentage (%)')

# Add legends
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(0.7, 1.0))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('emotic_radiant_fog_160_dev.csv')

# Count the occurrences of each category and calculate percentages
category_counts = df['category'].value_counts()
category_percentages = df['category'].value_counts(normalize=True) * 100

# Create a subplot with a shared x-axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the count numbers and percentages
bar = category_counts.plot(kind='bar', color='skyblue', label='Count', ax=ax)
for i, v in enumerate(category_counts):
    ax.text(i, v, f'{v}\n{category_percentages.iloc[i]:.2f}%', ha='center', va='bottom')

# Calculate the width for tighter layout
width = 0.4

# Plot the percentages on a secondary y-axis
ax2 = ax.twinx()
ax2.plot(ax.get_xticks(), category_percentages, color='orange', marker='o', label='Percentage (%)')

# Set x-axis labels with smaller distance between them
ax.set_xticks(range(len(category_counts)))
ax.set_xticklabels(category_counts.index, rotation=45, fontsize=10)

plt.title('Category Histogram')
ax.set_xlabel('Category')
ax.set_ylabel('Count')
ax2.set_ylabel('Percentage (%)')

# Add legends
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(0.7, 1.0))

plt.show()
```

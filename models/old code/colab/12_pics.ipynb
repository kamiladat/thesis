```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of samples
n_samples = 100

# Seed for reproducibility
np.random.seed(42)

# Generate random actual values for arousal and valence between 0 and 1
actual_arousal = np.random.uniform(low=0, high=1, size=n_samples)
actual_valence = np.random.uniform(low=0, high=1, size=n_samples)

# Generate predictions: actual values + noise based on MAE
# Assuming normal distribution of errors around the actual values
predicted_arousal = actual_arousal + np.random.normal(loc=0, scale=0.08, size=n_samples)
predicted_valence = actual_valence + np.random.normal(loc=0, scale=0.10, size=n_samples)

# Creating the plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for Arousal
ax[0].scatter(actual_arousal, predicted_arousal, alpha=0.6, color='blue')
ax[0].plot([0, 1], [0, 1], 'r--')  # Diagonal line for perfect prediction
ax[0].set_title('Arousal: Actual vs. Predicted')
ax[0].set_xlabel('Actual Arousal')
ax[0].set_ylabel('Predicted Arousal')

# Plot for Valence
ax[1].scatter(actual_valence, predicted_valence, alpha=0.6, color='green')
ax[1].plot([0, 1], [0, 1], 'r--')  # Diagonal line for perfect prediction
ax[1].set_title('Valence: Actual vs. Predicted')
ax[1].set_xlabel('Actual Valence')
ax[1].set_ylabel('Predicted Valence')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of samples
n_samples = 100

# Seed for reproducibility
np.random.seed(42)

# Generate random actual values for arousal and valence between -1 and 1
actual_arousal = np.random.uniform(low=-1, high=1, size=n_samples)
actual_valence = np.random.uniform(low=-1, high=1, size=n_samples)

# Generate predictions: actual values + noise based on MAE
# Assuming normal distribution of errors around the actual values
predicted_arousal = actual_arousal + np.random.normal(loc=0, scale=0.08, size=n_samples)
predicted_valence = actual_valence + np.random.normal(loc=0, scale=0.10, size=n_samples)

# Creating the plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for Arousal
ax[0].scatter(actual_arousal, predicted_arousal, alpha=0.6, color='blue')
ax[0].plot([-1, 1], [-1, 1], 'r--')  # Diagonal line for perfect prediction
ax[0].set_title('Arousal: Actual vs. Predicted')
ax[0].set_xlabel('Actual Arousal')
ax[0].set_ylabel('Predicted Arousal')
ax[0].set_xlim(-1, 1)
ax[0].set_ylim(-1, 1)

# Plot for Valence
ax[1].scatter(actual_valence, predicted_valence, alpha=0.6, color='green')
ax[1].plot([-1, 1], [-1, 1], 'r--')  # Diagonal line for perfect prediction
ax[1].set_title('Valence: Actual vs. Predicted')
ax[1].set_xlabel('Actual Valence')
ax[1].set_ylabel('Predicted Valence')
ax[1].set_xlim(-1, 1)
ax[1].set_ylim(-1, 1)

plt.tight_layout()
plt.show()
```

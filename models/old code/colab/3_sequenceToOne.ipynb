```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/AFEW-VA_radiant_fog_160_dev.csv')

# Extract video information from file paths
def extract_video_info(file_path):
    parts = file_path.split('/')
    video_id = parts[-2]  # Extract video ID from the file path
    return video_id

# Prepare video_frames dictionary
video_frames = {}
for idx, row in df.iterrows():
    file_path = row['path']
    video_id = extract_video_info(file_path)
    if video_id not in video_frames:
        video_frames[video_id] = []
    video_frames[video_id].append(file_path)

# Now video_frames contains a dictionary where each key is a video ID and its value is a list of frame paths

# Split frames into windows
window_size = 10
video_windows = {}
for video_id, frames in video_frames.items():
    windows = [frames[i:i + window_size] for i in range(0, len(frames), window_size)]
    video_windows[video_id] = windows
    print(windows)
    print("-----")

# Convert video_windows dictionary to DataFrame
window_data = {'video_id': [], 'window_frames': []}
for video_id, windows in video_windows.items():
    for window in windows:
        window_data['video_id'].append(video_id)
        window_data['window_frames'].append(window)

window_df = pd.DataFrame(window_data)

# Save the processed window DataFrame as a CSV file
window_df.to_csv('processed_video_windows.csv', index=False)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/AFEW-VA_radiant_fog_160_dev.csv')

# Extract video information from file paths
def extract_video_info(file_path):
    parts = file_path.split('/')
    video_id = parts[-2]  # Extract video ID from the file path
    return video_id

# Prepare video_frames dictionary
video_frames = {}
for idx, row in df.iterrows():
    file_path = row['path']
    video_id = extract_video_info(file_path)
    if video_id not in video_frames:
        video_frames[video_id] = []
    video_frames[video_id].append(file_path)

# Now video_frames contains a dictionary where each key is a video ID and its value is a list of frame paths

# Split frames into windows with overlap
window_size = 10
stride = 5  # The number of frames to move the window by
video_windows = {}
for video_id, frames in video_frames.items():
    windows = []
    for i in range(0, len(frames) - window_size + 1, stride):
        window = frames[i:i + window_size]
        windows.append(window)
    video_windows[video_id] = windows
    print(windows)
    print("-----")

# Convert video_windows dictionary to DataFrame
window_data = {'video_id': [], 'window_frames': []}
for video_id, windows in video_windows.items():
    for window in windows:
        window_data['video_id'].append(video_id)
        window_data['window_frames'].append(window)

window_df = pd.DataFrame(window_data)

# Save the processed window DataFrame as a CSV file
window_df.to_csv('processed_video_windows_with_stride.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv('SEWA_radiant_fog_160_train.csv')
dev_df = pd.read_csv('SEWA_radiant_fog_160_dev.csv')
test_df = pd.read_csv('SEWA_radiant_fog_160_test.csv')

# Function to plot distributions and print statistics
def plot_and_describe(df, dataset_name):
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Arousal distribution
    ax[0].hist(df['arousal'], bins=50, color='skyblue', edgecolor='black')
    ax[0].set_title(f'Distribution of Arousal Scores in {dataset_name}')
    ax[0].set_xlabel('Arousal Score')
    ax[0].set_ylabel('Frequency')

    # Valence distribution
    ax[1].hist(df['valence'], bins=50, color='salmon', edgecolor='black')
    ax[1].set_title(f'Distribution of Valence Scores in {dataset_name}')
    ax[1].set_xlabel('Valence Score')
    ax[1].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
    
    # Display basic statistics
    print(f"Basic Statistics for {dataset_name}:\n", df[['arousal', 'valence']].describe(), "\n\n")

# Analyze each dataset
plot_and_describe(train_df, "Training")
plot_and_describe(dev_df, "Development")
plot_and_describe(test_df, "Testing")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load datasets
train_df = pd.read_csv('SEWA_radiant_fog_160_train.csv')
dev_df = pd.read_csv('SEWA_radiant_fog_160_dev.csv')
test_df = pd.read_csv('SEWA_radiant_fog_160_test.csv')

# Combine the datasets for plotting
combined_df = pd.concat([train_df, dev_df, test_df])

# Find the range of the data to set plot limits
arousal_range = (combined_df['arousal'].min(), combined_df['arousal'].max())
valence_range = (combined_df['valence'].min(), combined_df['valence'].max())

# Plot arousal vs valence
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data points
ax.scatter(combined_df['arousal'], combined_df['valence'], alpha=0.5)

# Set the aspect of the plot to equal, to keep the scale the same on both axes
ax.set_aspect('equal')

# Set the limits of the plot to the range of the data
ax.set_xlim(arousal_range[0], arousal_range[1])
ax.set_ylim(valence_range[0], valence_range[1])

# Label the axes
ax.set_xlabel('Arousal')
ax.set_ylabel('Valence')

# Add grid
ax.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv('AFEW-VA_radiant_fog_160_train.csv')
dev_df = pd.read_csv('AFEW-VA_radiant_fog_160_dev.csv')
test_df = pd.read_csv('AFEW-VA_radiant_fog_160_test.csv')

# Combine the datasets for plotting
combined_df = pd.concat([train_df, dev_df, test_df])

# Option to exclude outliers; change this to `False` if you want to include the entire dataset
exclude_outliers = True

if exclude_outliers:
    # Calculate the quantiles
    low_quantile = combined_df.quantile(0.01)
    high_quantile = combined_df.quantile(0.99)
    
    # Filter out outliers beyond the 1st and 99th percentile
    combined_df = combined_df[(combined_df['arousal'] > low_quantile['arousal']) &
                              (combined_df['arousal'] < high_quantile['arousal']) &
                              (combined_df['valence'] > low_quantile['valence']) &
                              (combined_df['valence'] < high_quantile['valence'])]

# Plot arousal vs valence
fig, ax = plt.subplots(figsize=(6, 6))  # Using a square figure size

# Plot the data points
ax.scatter(combined_df['arousal'], combined_df['valence'], alpha=0.5)

# Set the aspect of the plot to equal, to keep the scale the same on both axes
ax.set_aspect('equal')

# Set the limits of the plot based on the filtered data range or set manually
# If you still want to use -10 to 10, uncomment the following two lines
ax.set_xlim(-10, 10)
# ax.set_ylim(-10, 10)

# If you want the limits to be set dynamically based on the filtered data, uncomment these:
# ax.set_xlim(combined_df['arousal'].min(), combined_df['arousal'].max())
# ax.set_ylim(combined_df['valence'].min(), combined_df['valence'].max())

# Label the axes
ax.set_xlabel('Arousal')
ax.set_ylabel('Valence')

# Add grid
ax.grid(True)

# Show the plot
plt.show()
```
